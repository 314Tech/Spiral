# Solution

The logarithmic spiral is a spiral whose polar equation is given by

\begin{align}
 r=ae^{b\theta} 	
\end{align}

where r is the distance from the origin, theta is the angle from the x-axis, and a and b are arbitrary constants. The logarithmic spiral is also known as the growth spiral, equiangular spiral, and spira mirabilis. It can be expressed parametrically as:

\begin{align}
x	=	rcos\theta = acos{\theta}e^{b\theta}	\\
y	=	rsin\theta = asin{\theta}e^{b\theta}
\end{align}

From the example above, the final point of the spiral is at angle 
\begin{align}
4\pi
\end{align}

```python
import matplotlib.pyplot as plt
```

It's at the last angle that the radius is our function parameter so:
\begin{align}
size	=	rcos\theta = acos{\theta}e^{b\theta}
\end{align}

To simplify, let's set b to -0.3 to get a similar direction than our example which gives us:
\begin{align}
 size=ae^{-0.3*4\pi} 	
\end{align}

And thus 
\begin{align}
 a=\frac{size}{e^{-0.3*4\pi}}	
\end{align}



```python
def draw_spiral(size):
    # Create the cartesian coordinates of the spiral from a series of angles
    max_angle = 4*math.pi
    step = max_angle/100
    a=size/math.exp(-0.3*4*math.pi)
    b=-0.3
    theta = 0
    x = []
    y = []
    fig, ax = plt.subplots()
    fig.set_size_inches(18.5, 10.5)
    while theta < max_angle:
        x.append(a*math.cos(theta)*math.exp(b*theta))
        y.append(a*math.sin(theta)*math.exp(b*theta))
        ax.plot(x, y)
        ax.set_aspect('equal')
        ax.grid(True, which='both')

        ax.axhline(y=0, color='k')
        ax.axvline(x=0, color='k')        
        theta += step
    plt.show()
```

```python
draw_spiral(1)
```

```python
draw_spiral(5)
```

```python
draw_spiral(10)
```
